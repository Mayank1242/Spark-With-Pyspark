```python
emp_df= spark.createDataFrame(emp_data,emp_schema)
```

```python
emp_df.show()
```

```python
emp_df.printSchema()
```

```python
## CastingColumn
# Casting means converting a column from one data type to another data type.
# Using select() (Not Recommended for Single Column Change)
# If transforming everything → use select()

from pyspark.sql.functions import col, cast
emp_casted =emp_df.select("employee_id","name","age",col("salary").cast("double"))

# With this method You are recreating the entire DataFrame
# You must explicitly mention all columns
# If you forget one → it will be dropped
```

```python
emp_casted.printSchema()
emp_casted.show()
```

```python
# Use withColumn() when modifying only one column:
# withColumn("column_name", expression)

from pyspark.sql.functions import col

emp_casted_one_col = emp_df.withColumn("salary", col("salary").cast("double"))
```

```python
emp_casted_one_col.show(5)
```

```python
# Adding Column on emp_casted
# now we are calucating the tax from the salary 
# withColumn() is used to:

# 1️ Modify an existing column
# 2️ Add a new column
# 3️ Create calculated columns

# Here with column will use to add the column

emp_taxed = emp_casted.withColumn("tax", col("salary") * 0.2)
emp_taxed.show()
```

```python
 # Literals
# In Spark, literal means a fixed constant value.
from pyspark.sql.functions import lit

emp_lit= emp_taxed.withColumn("Country",lit('India')).withColumn("County_Code",lit(91))
emp_lit.show()
```

```python
#Rename the coloumn
#Use withColumnRenamed

emp_lit = emp_lit.withColumnRenamed("employee_id","emp_id")
emp_lit.show()
```

```python
#Coloumn name with spaces
#But Space not recomanded in Project because maybe Downstream data will have Issue
emp_lit_withspace = emp_lit.withColumnRenamed("emp_id","emp id")
emp_lit_withspace.show()
```

```python
#Remove Column

#To delete multiple Column
# emp_dropped=emp_lit.drop("County_Code", "Country")
emp_dropped=emp_lit.drop("County_Code")
emp_dropped.show()
```

```python
#Filter Data
emp_filtered = emp_dropped.where("tax>10000")
emp_filtered.show()
```

```python

#Limit data

emp_limit= emp_filtered.limit(5)

# With show we can use that but when we want to store use limit
emp_limit.show(2)
```

```python
#Adding multiple Columns
# Use dictnories
columns = {
    "tax" : col("salary") * 0.2 ,
    "oneNumber" : lit(1), 
    "columnTwo" : lit("two")
}
emp_final = emp_df.withColumns(columns)

emp_final.show(5)
```
